```python
count = 0
for file_name in os.listdir(json_dir):
    if file_name.endswith('.json'):
        file_path = os.path.join(json_dir, file_name)
        count = count + 1
        print(file_path)
        
        with open(file_path, 'r', encoding='utf-8') as file:
            data = json.load(file)

        processed_data = process_messages(data['messages'])
        extracted_data = extract_relevant_data(processed_data)
        data = [item for item in extracted_data if not item['content'].startswith("Reacted")]
        data2 = data[::-1]
        for message in data2:
            if message['sender_name'] == 'Mohit Kondur':
                message['sender_name'] = 'Output'
            else:
                message['sender_name'] = 'Input'
        time_interval = 100 * 60 * 1000

        updated_messages = []
        
        print("*************************************************")
#         print(data2)
        for i in range(len(data2) - 1):
#             print(data2[i])
            updated_messages.append(data2[i])
            if (data2[i]['sender_name'] == 'Output' and 
                data2[i+1]['sender_name'] == 'Output' and 
                data2[i+1]['timestamp_ms'] - data2[i]['timestamp_ms'] > time_interval):                
                engagement_message = {
                    'sender_name': 'Input',
                    'timestamp_ms': data2[i+1]['timestamp_ms'] + 1, 
                    'content': 'Engagement.'
                }


                updated_messages.append(engagement_message)
                print("#########################################")
        if not updated_messages:
            continue 
                
        consolidate_input_output(updated_messages)

        if updated_messages[0]['sender_name'] =='Output':
            d = {
                'sender_name': 'Input',
                'timestamp_ms': updated_messages[0]['timestamp_ms']-10,
                'content': 'Engagement'}

            updated_messages.insert(0, d)

        if updated_messages[-1]['sender_name'] =='Input':
            updated_messages.pop()

        df = pd.DataFrame(updated_messages)

        df.drop(['timestamp_ms'], axis = 1, inplace = True)
        

        df['pseudo_index'] = (df['sender_name'] == 'Input').cumsum()
        

        final = df.pivot(index='pseudo_index', columns='sender_name', values='content')

        final.reset_index(drop = True, inplace = True)
        

        final.columns = ['Input', 'Output']
        main_df = pd.concat([main_df, final], ignore_index=True)
```
